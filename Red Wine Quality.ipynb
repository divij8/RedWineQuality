## As per the 75%ile and histogram, a good quality wine can be > 6

```python
x = df.iloc[:, :-1].values
y = df.loc[: , 'quality'].values
type(x)
```

```python
#y.value_counts()
#x.values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x , y , test_size = 0.2, random_state = 42)
```

```python
#Y_test.value_counts()
```

## Split is good, builiding model.

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error , r2_score
```

```python
regr = linear_model.LinearRegression()
regr.fit(X_train, Y_train)
```

```python
abc = df.columns
df.columns
abc = abc[0:-1]

coeff = pd.DataFrame(regr.coef_, abc, columns = ['Coefficients'])
coeff
```

```python
Y_pred = regr.predict(X_test)
```

```python
x = np.sqrt(mean_squared_error(Y_test, Y_pred))
x
```
